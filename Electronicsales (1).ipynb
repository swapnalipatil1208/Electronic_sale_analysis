```python
import pandas as pd
import plotly.express as px
from dash import Dash, dcc, html, Input, Output


df["Purchase Date"] = pd.to_datetime(df["Purchase Date"])
df.dropna(inplace=True)
df["Purchase Date"] = pd.to_datetime(df["Purchase Date"])
df["Year"] = df["Purchase Date"].dt.year
df["Month"] = df["Purchase Date"].dt.month
df["Year-Month"] = df["Purchase Date"].dt.to_period("M").astype(str)
df["Price Range"] = pd.cut(df["Product Price"], bins=10)


app = Dash(__name__)
app.title = "Electronic Sales Dashboard"

app.layout = html.Div([
    html.H1("📈 Electronic Sales Analysis", style={"textAlign": "center"}),

    html.Div([
        html.Label("Select Year(s):"),
        dcc.Dropdown(
            id="year-filter",
            options=[{"label": str(year), "value": year} for year in sorted(df["Year"].unique())],
            value=sorted(df["Year"].unique()),
            multi=True,
            style={"width": "50%"}
        )
    ], style={"padding": "20px"}),

    html.Div(id="kpi-cards", style={"display": "flex", "justifyContent": "space-around", "marginBottom": "40px"}),

    html.Div([
        dcc.Graph(id="category-sales", style={"width": "48%"}),
        dcc.Graph(id="yearly-sales", style={"width": "48%"}),
        dcc.Graph(id="payment-method", style={"width": "48%"}),
        dcc.Graph(id="top-products", style={"width": "48%"}),
        dcc.Graph(id="price-distribution", style={"width": "48%"}),
        dcc.Graph(id="monthly-trend", style={"width": "48%"}),
        dcc.Graph(id="correlation-heatmap", style={"width": "48%"}),
        dcc.Graph(id="avg-quantity", style={"width": "48%"}),
        dcc.Graph(id="growth-rate", style={"width": "48%"}),
    ], style={
        "display": "flex",
        "flexWrap": "wrap",
        "justifyContent": "space-between",
        "gap": "20px",
        "padding": "0 20px 40px"
    }),
])

@app.callback(
    Output("kpi-cards", "children"),
    Output("category-sales", "figure"),
    Output("yearly-sales", "figure"),
    Output("payment-method", "figure"),
    Output("top-products", "figure"),
    Output("price-distribution", "figure"),
    Output("monthly-trend", "figure"),
    Output("correlation-heatmap", "figure"),
    Output("avg-quantity", "figure"),
    Output("growth-rate", "figure"),
    Input("year-filter", "value")
)
def update_dashboard(selected_years):
    dff = df[df["Year"].isin(selected_years)]

    
    total_orders = dff["Order ID"].nunique()
    total_revenue = dff["Total Sale"].sum()
    total_units = dff["Unit Sold"].sum()

    yearly = dff.groupby("Year").agg({
        "Order ID": "nunique",
        "Total Sale": "sum",
        "Unit Sold": "sum"
    }).reset_index()

    def make_kpi_card(title, value, breakdown_label, breakdown_values):
        return html.Div([
            html.H4(title),
            html.H2(f"{value:,}"),
            html.Ul([html.Li(f"{row['Year']}: {breakdown_values(row)}") for _, row in yearly.iterrows()])
        ], style={
            "padding": "20px",
            "border": "1px solid #ccc",
            "borderRadius": "10px",
            "textAlign": "center",
            "width": "30%",
            "boxShadow": "0 2px 6px rgba(0,0,0,0.1)"
        })

    kpi_cards = [
        make_kpi_card("🛒 Total Orders", total_orders, "Orders", lambda r: f"{int(r['Order ID'])}"),
        make_kpi_card("💰 Total Revenue", total_revenue, "Revenue", lambda r: f"₹{r['Total Sale']:,.0f}"),
        make_kpi_card("📦 Total Units Sold", total_units, "Units", lambda r: f"{int(r['Unit Sold'])}")
    ]

   
    cat_sales = dff.groupby("Category")["Total Sale"].sum().sort_values()
    fig1 = px.bar(cat_sales, x=cat_sales.values, y=cat_sales.index, orientation="h", title="Category-Wise Sales Distribution")

    
    yearly_sales = dff.groupby("Year")["Total Sale"].sum()
    fig2 = px.line(x=yearly_sales.index, y=yearly_sales.values, markers=True, title="Yearly Sales Trend")

   
    pay_counts = dff["Payment Method"].value_counts()
    fig3 = px.pie(values=pay_counts.values, names=pay_counts.index, title="Payment Method Distribution", hole=0.3)

   
    top10_df = dff.groupby("Product Name").agg({
        "Total Sale": "sum",
        "Unit Sold": "sum"
    }).sort_values(by="Total Sale", ascending=False).head(10).reset_index()

    fig4 = px.bar(
        top10_df,
        x="Total Sale",
        y="Product Name",
        orientation="h",
        text="Unit Sold",
        title="Top 10 Best-Selling Products (with Quantity Sold)",
        labels={"Total Sale": "Total Sales", "Product Name": "Product"}
    )
    fig4.update_traces(texttemplate="%{text:,} units", textposition="outside")
    fig4.update_layout(yaxis=dict(autorange="reversed"))

    
    fig5 = px.box(dff, x="Category", y="Product Price", title="Price Distribution by Category")

    
    monthly_sales = dff.groupby("Year-Month")["Total Sale"].sum()
    fig6 = px.line(x=monthly_sales.index, y=monthly_sales.values, markers=True, title="Sales Trend Over Time")

    
    corr = dff[["Product Price", "Unit Sold"]].corr()
    fig7 = px.imshow(corr, text_auto=True, title="Correlation Between Price and Quantity")

    
    avg_q = dff.groupby("Price Range")["Unit Sold"].mean()
    fig8 = px.bar(x=avg_q.index.astype(str), y=avg_q.values,
                  title="Average Quantity Sold by Price Range", labels={"x": "Price Range", "y": "Average Quantity"})

    
    monthly_growth = monthly_sales.pct_change().fillna(0) * 100
    fig9 = px.line(x=monthly_growth.index, y=monthly_growth.values, markers=True,
                   title="Monthly Sales Growth Rate (%)")

    return kpi_cards, fig1, fig2, fig3, fig4, fig5, fig6, fig7, fig8, fig9

if __name__ == "__main__":
    app.run(debug=True, port=8052)
    
```
